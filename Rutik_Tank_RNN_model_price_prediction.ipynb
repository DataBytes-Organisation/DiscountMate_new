### Concate all data with dates

```python
# Add a date column to each dataframe
df_1['Date'] = '2024-08-04'
df_2['Date'] = '2024-09-03'
df_3['Date'] = '2024-08-27'
df_4['Date'] = '2024-09-04'

# Concatenate the dataframes
df_combined = pd.concat([df_1, df_2, df_3, df_4], ignore_index=True)

df_combined.head()
```

### Analyzing Price Trends Over Time


```python
import matplotlib.pyplot as plt

# Convert price columns to numeric values
df_combined['Best Price'] = df_combined['Best Price'].replace('[\$,]', '', regex=True).astype(float)
df_combined['Item Price'] = df_combined['Item Price'].replace('[\$,]', '', regex=True).astype(float)

# Group by Date and calculate average prices
price_trends = df_combined.groupby('Date')[['Best Price', 'Item Price']].mean().reset_index()

# Plot the trends
plt.figure(figsize=(12, 6))
plt.plot(price_trends['Date'], price_trends['Best Price'], label='Best Price', marker='o')
plt.plot(price_trends['Date'], price_trends['Item Price'], label='Item Price', marker='X')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.title('Average Price Trends Over Time')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Histogram for Best Price
plt.figure(figsize=(10, 6))
sns.histplot(df_combined['Best Price'], bins=30, kde=True)
plt.title('Distribution of Best Price')
plt.xlabel('Best Price')
plt.ylabel('Frequency')
plt.show()

# Box Plot for Best Price
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Best Price', data=df_combined)
plt.title('Box Plot of Best Price by Category')
plt.xlabel('Category')
plt.ylabel('Best Price')
plt.xticks(rotation=45)
plt.show()
```

## Category-wise Bar Chart:


```python
# Group by Category and calculate average prices
category_avg_prices = df_combined.groupby('Category')[['Best Price', 'Item Price']].mean().reset_index()

# Bar Chart for Average Prices by Category
plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Best Price', data=category_avg_prices, label='Best Price', color='b')
sns.barplot(x='Category', y='Item Price', data=category_avg_prices, label='Item Price', color='r')
plt.title('Average Prices by Category')
plt.xlabel('Category')
plt.ylabel('Average Price')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

## Price prediction VAR Model

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM

# Assuming df_combined is your combined dataframe
# Convert 'Item Price' to numeric
df_combined['Item Price'] = pd.to_numeric(df_combined['Item Price'], errors='coerce')

# Drop rows with NaN values in 'Item Price'
df_combined.dropna(subset=['Item Price'], inplace=True)

# Normalize the 'Item Price' column
scaler = MinMaxScaler(feature_range=(0, 1))
df_combined['Item Price'] = scaler.fit_transform(df_combined['Item Price'].values.reshape(-1, 1))

# Prepare the data for RNN
def create_dataset(data, time_step=1):
    X, Y = [], []
    for i in range(len(data)-time_step-1):
        a = data[i:(i+time_step), 0]
        X.append(a)
        Y.append(data[i + time_step, 0])
    return np.array(X), np.array(Y)

time_step = 10
data = df_combined['Item Price'].values
data = data.reshape(-1, 1)
X, Y = create_dataset(data, time_step)

# Reshape input to be [samples, time steps, features]
X = X.reshape(X.shape[0], X.shape[1], 1)

# Create the RNN model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(time_step, 1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X, Y, epochs=25, batch_size=32, verbose=1)

# Predicting the prices
predicted_prices = model.predict(X)
predicted_prices = scaler.inverse_transform(predicted_prices)

# Evaluate the model
from sklearn.metrics import mean_squared_error
import math

rmse = math.sqrt(mean_squared_error(Y, predicted_prices))
print(f'Root Mean Squared Error: {rmse}')
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

mae = mean_absolute_error(Y, predicted_prices)

# Custom accuracy function
def accuracy(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

accuracy_value = accuracy(Y, predicted_prices)

print(f'Root Mean Squared Error: {rmse}')
print(f'Mean Absolute Error: {mae}')
print(f'Accuracy: {100 - accuracy_value}%')
```
